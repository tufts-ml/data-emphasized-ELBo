```python
import os
import math
import itertools
import numpy as np
import pandas as pd
# PyTorch
import torch
import torchvision
```

```python
def check_epochs(df, n, batch_size=128, steps=6000):
    num_batches = math.floor(n / batch_size)
    epochs = int(steps / num_batches)
    return True if df.shape[0] == epochs else False
```

```python
model = torchvision.models.resnet50()
model.fc = torch.nn.Linear(in_features=2048, out_features=10, bias=True)
print(f"d = {len(torch.nn.utils.parameters_to_vector(model.parameters()))}")
model = torchvision.models.vit_b_16()
model.heads = torch.nn.Linear(in_features=768, out_features=10, bias=True)
print(f"d = {len(torch.nn.utils.parameters_to_vector(model.parameters()))}")
model = torchvision.models.convnext_tiny()
model.classifier[2] = torch.nn.Linear(in_features=768, out_features=10, bias=True)
print(f"d = {len(torch.nn.utils.parameters_to_vector(model.parameters()))}")
```

```python
criterion = "l2-sp"
d = 23528522
dataset = "CIFAR-10"
dataset_dir = "/cluster/tufts/hugheslab/eharve06/CIFAR-10"
experiments_dir = "/cluster/tufts/hugheslab/eharve06/retrained_CIFAR-10_diagVI"
lr_0s = [0.1, 0.01, 0.001, 0.0001]
model_arch = "ResNet-50"
ns = [100, 1000, 10000, 50000]
prior_dir = "/cluster/tufts/hugheslab/eharve06/resnet50_torchvision"
prior_type = "resnet50_torchvision"
random_states = [1001, 2001, 3001]

count = -1
for lr_0, n, random_state in itertools.product(lr_0s, ns, random_states):
    model_name = f"{criterion}_kappa={d/n}_lr_0={lr_0}_n={n}_random_state={random_state}"
    if os.path.exists(f"{experiments_dir}/{model_name}.csv"):
        temp_df = pd.read_csv(f"{experiments_dir}/{model_name}.csv")
        if not check_epochs(temp_df, n, batch_size=min(128, n), steps=6000):
            count += 1
            print(f"    'python ../src/main_image_classifiers.py --batch_size=128 --dataset=\"{dataset}\" --dataset_dir=\"{dataset_dir}\" --experiments_dir=\"{experiments_dir}\" --kappa={d/n} --lr_0={lr_0} --model_arch=\"{model_arch}\" --model_name=\"{model_name}\" --n={n} --num_workers=0 --prior_dir=\"{prior_dir}\" --prior_type=\"{prior_type}\" --random_state={random_state} --save'")
    else:
        count += 1
        print(f"    'python ../src/main_image_classifiers.py --batch_size=128 --dataset=\"{dataset}\" --dataset_dir=\"{dataset_dir}\" --experiments_dir=\"{experiments_dir}\" --kappa={d/n} --lr_0={lr_0} --model_arch=\"{model_arch}\" --model_name=\"{model_name}\" --n={n} --num_workers=0 --prior_dir=\"{prior_dir}\" --prior_type=\"{prior_type}\" --random_state={random_state} --save'")
print(count)
```

```python
criterion = "l2-sp"
d = 23528522
#d = 85806346
#d = 27827818
experiments_dir = "/cluster/tufts/hugheslab/eharve06/retrained_CIFAR-10_diagVI"
lr_0s = [0.1, 0.01, 0.001, 0.0001]
ns = [100, 1000, 10000, 50000]
random_states = [1001, 2001, 3001]

columns = ["criterion", "model_name", "n", "train_loss", "test_acc", "test_nll", "random_state", "runtime"]
cifar10_deelbo_diag_tuned_df = pd.DataFrame(columns=columns)

for lr_0, n, random_state in itertools.product(lr_0s, ns, random_states):
    model_name = f"{criterion}_kappa={d/n}_lr_0={lr_0}_n={n}_random_state={random_state}"
    if os.path.exists(f"{experiments_dir}/{model_name}.csv"):
        temp_df = pd.read_csv(f"{experiments_dir}/{model_name}.csv")
        row = [criterion, model_name, n, temp_df.train_loss.values[-1], temp_df.val_or_test_acc.values[-1], temp_df.val_or_test_nll.values[-1], random_state, temp_df["train_sec/epoch"].sum()]
        cifar10_deelbo_diag_tuned_df.loc[len(cifar10_deelbo_diag_tuned_df)] = row
    
min_indices = cifar10_deelbo_diag_tuned_df.groupby(["criterion", "n", "random_state"])["train_loss"].idxmin()
cifar10_deelbo_diag_retrained_df = cifar10_deelbo_diag_tuned_df.loc[min_indices]
cifar10_deelbo_diag_retrained_df["total_time"] = cifar10_deelbo_diag_tuned_df.groupby(["criterion", "n", "random_state"])["runtime"].sum().values
```

```python
grouped_df = cifar10_deelbo_diag_retrained_df.groupby(['criterion', 'n']).agg(lambda x: tuple(x))
columns = ['test_acc', 'test_nll']
for column in columns:
    grouped_df[f'{column}_mean'] = grouped_df[column].apply(lambda item: np.mean(item))
    grouped_df[f'{column}_std'] = grouped_df[column].apply(lambda item: np.std(item))
    grouped_df[f'{column}_min'] = grouped_df[column].apply(lambda item: np.min(item))
    grouped_df[f'{column}_max'] = grouped_df[column].apply(lambda item: np.max(item))
grouped_df = grouped_df.reset_index()
grouped_df
```
