```python
experiments_directory = '/cluster/tufts/hugheslab/eharve06/data-emphasized-ELBo/experiments/train_eval_comparison'
first_model_name = 'l2-sp_kappa=1.0_lr_0=0.01_n=1000_random_state=1001'
second_model_name = 'l2-sp_kappa=23528.522_lr_0=0.01_n=1000_random_state=1001'
third_model_name = 'l2-sp_kappa=1.0_lr_0=0.001_n=1000_random_state=1001'
fourth_model_name = 'l2-sp_kappa=23528.522_lr_0=0.001_n=1000_random_state=1001'
temp_df1 = pd.read_csv(f'{experiments_directory}/{first_model_name}.csv')
temp_df2 = pd.read_csv(f'{experiments_directory}/{second_model_name}.csv')
temp_df3 = pd.read_csv(f'{experiments_directory}/{third_model_name}.csv')
temp_df4 = pd.read_csv(f'{experiments_directory}/{fourth_model_name}.csv')
```

```python
kappa, N = 23528.522, 10000

ncols, nrows = 2, 3
fig, axs = plt.subplots(figsize=(4*ncols, 3*nrows), ncols=ncols, nrows=nrows, sharex=True, sharey='row')

axs[0,0].plot(temp_df1.epoch, N * temp_df1.train_nll + temp_df1.train_kl, color='#E377C2', linewidth=3)
axs[0,0].plot(temp_df2.epoch, N * temp_df2.train_nll + temp_df2.train_kl, color='#9467BD', linewidth=3)
axs[0,0].set_yscale('log')
axs[0,0].set_title(r'$N=1000, \text{lr}=0.01, \text{seed}=1001$')
axs[0,0].set_ylabel('Negative ELBo')
axs[0,0].grid()

axs[0,1].plot(temp_df3.epoch, N * temp_df3.train_nll + temp_df3.train_kl, color='#E377C2', linewidth=3)
axs[0,1].plot(temp_df4.epoch, N * temp_df4.train_nll + temp_df4.train_kl, color='#9467BD', linewidth=3)
axs[0,1].set_yscale('log')
axs[0,1].set_title(r'$N=1000, \text{lr}=0.001, \text{seed}=1001$')
axs[0,1].grid()

axs[1,0].plot(temp_df1.epoch, temp_df1.train_kl, color='#E377C2', linewidth=3)
axs[1,0].plot(temp_df2.epoch, temp_df2.train_kl, color='#9467BD', linewidth=3)
axs[1,0].set_yscale('log')
axs[1,0].set_ylabel('KL')
axs[1,0].grid()

axs[1,1].plot(temp_df3.epoch, temp_df3.train_kl, color='#E377C2', linewidth=3)
axs[1,1].plot(temp_df4.epoch, temp_df4.train_kl, color='#9467BD', linewidth=3)
axs[1,1].set_yscale('log')
axs[1,1].grid()

axs[2,0].plot(temp_df1.epoch, N * temp_df1.train_nll, color='#E377C2', label='ELBo', linewidth=3)
axs[2,0].plot(temp_df2.epoch, N * temp_df2.train_nll, color='#9467BD', label='DE ELBo (ours)', linewidth=3)
axs[2,0].set_yscale('log')
axs[2,0].set_xlabel('Epoch')
axs[2,0].set_ylabel('Expected NLL')
axs[2,0].legend(loc='lower left')
axs[2,0].grid()

axs[2,1].plot(temp_df3.epoch, N * temp_df3.train_nll, color='#E377C2', linewidth=3)
axs[2,1].plot(temp_df4.epoch, N * temp_df4.train_nll, color='#9467BD', linewidth=3)
axs[2,1].set_yscale('log')
axs[2,1].set_xlabel('Epoch')
axs[2,1].grid()

fig.tight_layout()
#fig.savefig('train_eval_comparison.pdf', bbox_inches='tight')
plt.show()
```
