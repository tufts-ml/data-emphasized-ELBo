```python
import os
import math
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 11})
import matplotlib.ticker as ticker
# PyTorch
import torch
import torchvision
```

```python
def check_epochs(df, n, batch_size=128, steps=6000):
    num_batches = math.floor(n/batch_size)
    epochs = int(steps/num_batches)
    if df.shape[0] == epochs:
        return True
    else:
        return False
```

```python
criterion = 'l2-sp'
dataset_directory = '/cluster/tufts/hugheslab/eharve06/CIFAR-10'
experiments_directory = '/cluster/tufts/hugheslab/eharve06/data-emphasized-ELBo/experiments/retrained_CIFAR-10_ConvNeXt_Tiny_VI'
kappas = torch.logspace(0, math.log10(27827818), 10)[1:]
lr_0s = [0.1, 0.01, 0.001, 0.0001]
model_arch = 'ConvNeXt_Tiny'
ns = [100, 1000, 10000, 50000]
prior_directory = '/cluster/tufts/hugheslab/eharve06/convnext_tiny_torchvision'
prior_type = 'convnext_tiny_torchvision'
random_states = [1001, 2001, 3001]

count = -1
for kappa, lr_0, n, random_state in itertools.product(kappas, lr_0s, ns, random_states):
    model_name = f'{criterion}_kappa={kappa}_lr_0={lr_0}_n={n}_random_state={random_state}'
    if os.path.exists(f'{experiments_directory}/{model_name}.csv'):
        temp_df = pd.read_csv(f'{experiments_directory}/{model_name}.csv')
        if not check_epochs(temp_df, n, batch_size=min(128, n), steps=6000):
            count += 1
            print(f'    "python ../src/main_CIFAR-10.py --batch_size=128 --criterion=\'{criterion}\' --dataset_directory=\'{dataset_directory}\' --ELBo --experiments_directory=\'{experiments_directory}\' --kappa={kappa} --lr_0={lr_0} --model_arch=\'{model_arch}\' --model_name=\'{model_name}\' --n={n} --num_workers=0 --prior_directory=\'{prior_directory}\' --prior_type=\'{prior_type}\' --random_state={random_state} --save"')
    else:
        count += 1
        print(f'    "python ../src/main_CIFAR-10.py --batch_size=128 --criterion=\'{criterion}\' --dataset_directory=\'{dataset_directory}\' --ELBo --experiments_directory=\'{experiments_directory}\' --kappa={kappa} --lr_0={lr_0} --model_arch=\'{model_arch}\' --model_name=\'{model_name}\' --n={n} --num_workers=0 --prior_directory=\'{prior_directory}\' --prior_type=\'{prior_type}\' --random_state={random_state} --save"')
print(count)
```

```python
criterion = 'l2-sp'
dataset_directory = '/cluster/tufts/hugheslab/eharve06/CIFAR-10'
experiments_directory = '/cluster/tufts/hugheslab/eharve06/data-emphasized-ELBo/experiments/retrained_CIFAR-10_ConvNeXt_Tiny_VI'
kappas = torch.logspace(0, math.log10(27827818), 10)
lr_0s = [0.1, 0.01, 0.001, 0.0001]
ns = [100, 1000, 10000, 50000]
random_states = [1001, 2001, 3001]

columns = ['criterion', 'kappa', 'model_name', 'n', 'random_state', 'train_loss', 'test_acc', 'test_nll']
cifar10_retrained_df = pd.DataFrame(columns=columns)

for kappa, lr_0, n, random_state in itertools.product(kappas, lr_0s, ns, random_states):
    model_name = f'{criterion}_kappa={kappa}_lr_0={lr_0}_n={n}_random_state={random_state}'
    if os.path.exists(f'{experiments_directory}/{model_name}.csv'):
        temp_df = pd.read_csv(f'{experiments_directory}/{model_name}.csv')
        row = [criterion, kappa.item(), model_name, n, random_state, temp_df.train_loss.values[-1], temp_df.val_or_test_acc.values[-1], temp_df.val_or_test_nll.values[-1]]
        cifar10_retrained_df.loc[len(cifar10_retrained_df)] = row
        
for lr_0, n, random_state in itertools.product(lr_0s, ns, random_states):
    model_name = f'{criterion}_kappa={27827818/n}_lr_0={lr_0}_n={n}_random_state={random_state}'
    if os.path.exists(f'{experiments_directory}/{model_name}.csv'):
        temp_df = pd.read_csv(f'{experiments_directory}/{model_name}.csv')
        row = [criterion, 27827818/n, model_name, n, random_state, temp_df.train_loss.values[-1], temp_df.val_or_test_acc.values[-1], temp_df.val_or_test_nll.values[-1]]
        cifar10_retrained_df.loc[len(cifar10_retrained_df)] = row
    
min_indices = cifar10_retrained_df.groupby(['criterion', 'kappa', 'n', 'random_state'])['train_loss'].idxmin()
cifar10_retrained_df = cifar10_retrained_df.loc[min_indices]

cifar10_retrained_df
```

```python
cifar10_grouped_df = cifar10_retrained_df.groupby(['criterion', 'kappa', 'n']).agg(lambda x: tuple(x))
columns = ['test_acc', 'test_nll']
for column in columns:
    cifar10_grouped_df[f'{column}_mean'] = cifar10_grouped_df[column].apply(lambda item: np.mean(item))
    cifar10_grouped_df[f'{column}_std'] = cifar10_grouped_df[column].apply(lambda item: np.std(item))
    cifar10_grouped_df[f'{column}_min'] = cifar10_grouped_df[column].apply(lambda item: np.min(item))
    cifar10_grouped_df[f'{column}_max'] = cifar10_grouped_df[column].apply(lambda item: np.max(item))
cifar10_grouped_df = cifar10_grouped_df.reset_index()
cifar10_grouped_df
```
